```python
# -------------------------------
# Expense Tracker System
# Python Mini Project
# -------------------------------

expenses =[{"Date": "02-01-2026", "Category": "FOOD", "Amount": 250}]
 # List to store expense records


def add_expense():
    try:
        date = input("Enter Date (DD-MM-YYYY): ").strip()
        category = input("Enter Category: ").strip().upper()
        amount = int(input("Enter Amount: "))

        expense = {
            "Date": date,
            "Category": category,
            "Amount": amount
        }

        expenses.append(expense)
        print("Expense added successfully!")

    except ValueError:
        print("Invalid input! Amount must be a number.")


def view_expenses():
    if len(expenses) == 0:
        print("No expenses recorded.")
    else:
        print("\n--- Expense Records ---")
        for exp in expenses:
            print("Date:", exp.get("Date"))
            print("Category:", exp.get("Category"))
            print("Amount:", exp.get("Amount"))
            print("----------------------")


def search_expense():
    search_category = input("Enter category to search: ").strip().upper()

    found = False
    for exp in expenses:
        if exp.get("Category") == search_category:
            print(exp)
            found = True

    if not found:
        print("No expense found for this category.")


def save_to_file():
    with open("expenses.txt", "w") as file:
        for exp in expenses:
            file.write(str(exp) + "\n")
    print("Expenses saved to file.")


def menu():
    while True:
        print("\n===== Expense Tracker Menu =====")
        print("1. Add Expense")
        print("2. View Expenses")
        print("3. Search Expense by Category")
        print("4. Save & Exit")

        choice = input("Enter your choice (1-4): ").strip()

        if choice == "1":
            add_expense()
        elif choice == "2":
            view_expenses()
        elif choice == "3":
            search_expense()
        elif choice == "4":
            save_to_file()
            print("Exiting program. Thank you!")
            break
        else:
            print("Invalid choice. Please select 1 to 4.")


# Program starts here
menu()
```
